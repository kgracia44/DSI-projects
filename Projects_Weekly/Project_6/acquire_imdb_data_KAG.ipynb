```python
results = []


for title in idf['title']:
    title = title.encode('utf8')
    title = title.replace(" ", "+")
    URL = "http://www.omdbapi.com/?t={title}&plot=full&r=xml&tomatoes=true".format(**locals())
    i = requests.get(URL)
    i_soup = BeautifulSoup(i.content, "lxml-xml")
    results.append(i_soup)
    
# results
```

```python
def extract_everything(list_of_soup_files):
    """Extract desired movie info using BS4. Future: will refactor using modulo (%) 
    to pass in a list of desired tag/attribute names if possible..."""
    list_dict_250 = []
    for soup_file in list_of_soup_files:
        dict_250 = {}
        
        titles=soup_file.findAll('movie',{"title":True})
        for title in titles:
            dict_250['title'] = title['title']
        
        ratings=soup_file.findAll('movie',{"rated":True})
        for rating in ratings:
            dict_250['rating'] = str(rating['rated'])
        
        release_dates=soup_file.findAll('movie',{"released":True})
        for release in release_dates:
            dict_250['released'] = str(release['released'])
        
        run_times=soup_file.findAll('movie',{"runtime":True})
        for run_time in run_times:
            dict_250['runtime'] = str(run_time['runtime'])
        
        genres=soup_file.findAll('movie',{"genre":True})
        for genre in genres:
            dict_250['genre'] = str(genre['genre'])
        
        directors=soup_file.findAll('movie',{"director":True})
        for director in directors:
            dict_250['director'] = director['director'].encode('utf8')
        
        writers=soup_file.findAll('movie',{"writer":True})
        for writer in writers:
            dict_250['writer'] = writer['writer'].encode('utf8')
        
        actors=soup_file.findAll('movie',{"actors":True})
        for actor in actors:
            dict_250['actors'] = actor['actors'].encode('utf8')
        
        plots=soup_file.findAll('movie',{"plot":True})
        for plot in plots:
            dict_250['plot'] = plot['plot'].encode('utf8')
        
        languages=soup_file.findAll('movie',{"language":True})
        for language in languages:
            dict_250['language'] = str(language['language'])
        
        countrys=soup_file.findAll('movie',{"country":True})
        for country in countrys:
            dict_250['country'] = str(country['country'])
        
        awards=soup_file.findAll('movie',{"awards":True})
        for award in awards:
            dict_250['awards'] = str(award['awards'])
        
        metascores=soup_file.findAll('movie',{"metascore":True})
        for metascore in metascores:
            dict_250['metascore'] = str(metascore['metascore'])
        
        imdbRatings=soup_file.findAll('movie',{"imdbRating":True})
        for imdbRating in imdbRatings:
            dict_250['imdbRating'] = str(imdbRating['imdbRating'])
        
        imdbVotes=soup_file.findAll('movie',{"imdbVotes":True})
        for imdbVote in imdbVotes:
            dict_250['imdbVotes'] = str(imdbVote['imdbVotes'])
        
        imdbIDs=soup_file.findAll('movie',{"imdbID":True})
        for imdbID in imdbIDs:
            dict_250['imdbID'] = str(imdbID['imdbID'])
        
        tomatoMeters=soup_file.findAll('movie',{"tomatoMeter":True})
        for tomatoMeter in tomatoMeters:
            dict_250['tomatoMeter'] = str(tomatoMeter['tomatoMeter'])
    
        tomatoImages=soup_file.findAll('movie',{"tomatoImage":True})
        for tomatoImage in tomatoImages:
            dict_250['tomatoImage'] = str(tomatoImage['tomatoImage'])
                
        tomatoRatings=soup_file.findAll('movie',{"tomatoRating":True})
        for tomatoRating in tomatoRatings:
            dict_250['tomatoRating'] = str(tomatoRating['tomatoRating'])
                
        tomatoReviews=soup_file.findAll('movie',{"tomatoReviews":True})
        for tomatoReview in tomatoReviews:
            dict_250['tomatoReviews'] = str(tomatoReview['tomatoReviews'])
                
        tomatoFreshs=soup_file.findAll('movie',{"tomatoFresh":True})
        for tomatoFresh in tomatoFreshs:
            dict_250['tomatoFresh'] = str(tomatoFresh['tomatoFresh'])
                
        tomatoRottens=soup_file.findAll('movie',{"tomatoRotten":True})
        for tomatoRotten in tomatoRottens:
            dict_250['tomatoRotten'] = str(tomatoRotten['tomatoRotten'])
                
        tomatoConsensuses=soup_file.findAll('movie',{"tomatoConsensus":True})
        for tomatoConsensus in tomatoConsensuses:
            dict_250['tomatoConsensus'] = tomatoConsensus['tomatoConsensus'].encode('utf8')
                
        tomatoUserMeters=soup_file.findAll('movie',{"tomatoUserMeter":True})
        for tomatoUserMeter in tomatoUserMeters:
            dict_250['tomatoUserMeter'] = tomatoUserMeter['tomatoUserMeter'].encode('utf8')
                
        tomatoUserRatings=soup_file.findAll('movie',{"tomatoUserRating":True})
        for tomatoUserRating in tomatoUserRatings:
            dict_250['tomatoUserRating'] = str(tomatoUserRating['tomatoUserRating'])
                
        tomatoUserReviews=soup_file.findAll('movie',{"tomatoUserReviews":True})
        for tomatoUserReview in tomatoUserReviews:
            dict_250['tomatoUserReviews'] = str(tomatoUserReview['tomatoUserReviews'])
                
        DVDs=soup_file.findAll('movie',{"DVD":True})
        for DVD in DVDs:
            dict_250['DVD'] = str(DVD['DVD'])
                
        BoxOffices=soup_file.findAll('movie',{"BoxOffice":True})
        for BoxOffice in BoxOffices:
            dict_250['BoxOffice'] = str(BoxOffice['BoxOffice'])
                
        Productions=soup_file.findAll('movie',{"Production":True})
        for Production in Productions:
            dict_250['Production'] = str(Production['Production'])
        
        list_dict_250.append(dict_250)        
    
    return list_dict_250

top_250_raw = extract_everything(results)
```
